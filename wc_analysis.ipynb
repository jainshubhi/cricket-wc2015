```python
df = pd.read_csv('world_cup_2015_games.csv')
```

```python
matches = df.match_id.values
ground_ids = df.ground_id.values
ground_names = df.ground_name.values
dates = df.date.values
```

```python
df = pd.DataFrame()
for i in xrange(len(matches)):
    match = matches[i]
    ground_id = ground_ids[i]
    ground_name = ground_names[i]
    date = dates[i]
    df2 = pd.read_csv('wc_cleaned_data/%s_summary.csv'%match)
    #df2['match_id'] = match
    #df2['ground_id'] = ground_id
    df2['ground_name'] = ground_name
    df2['date'] = date
    df = df.append(df2)
```

```python
def f(ball_speed):
    if ball_speed > 0:
        return ball_speed
    else:
        return np.nan
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
def left_hander_standardization(row):
    if row['bat_right_handed'] == 'n':
        return -1*row['ended_x'], -1*row['landing_x']
    else: return row['ended_x'], row['landing_x']
```

```python
df['ended_x'], df['landing_x'] = zip(*df.apply(left_hander_standardization, axis=1))
```

```python
df['dummy_ball'] = ~((df.extras_type == 'Wd') | (df.extras_type == 'Nb'))
df['dot_ball'] = df['runs_w_extras'] == 0
df['single'] = df['runs_batter'] == 1
df['double'] = df['runs_batter'] == 2
df['triple'] = df['runs_batter'] == 3
df['four'] = (df['runs_batter'] == 4) | (df['runs_batter'] == 5)
df['six'] = df['runs_batter'] == 6
```

```python
df['dummy_ball'] = df['dummy_ball'].astype(int)
```

```python
def get_height(i):
    if i <=0.5: return 'low_height'
    elif i <=1.0: return 'medium_height'
    elif i <=1.5: return 'high_height'
    else: return 'bouncer'
```

```python
def get_line(i):
    if i <-0.6: return 'way_outside_off'
    elif i <-0.3: return 'outside_off'
    elif i <-0.15: return 'just_outside_off'
    elif i < 0: return 'on_off_stump'
    elif i <0.15: return 'on_leg_stump'
    else: return 'outside_leg'
```

```python
df['ball_height'] = df['ended_y'].apply(get_height)
df['ball_ended_line'] = df['ended_x'].apply(get_line)
```

```python
df['bowler_wicket'] = (df['wicket_method'].notnull()) & (df['wicket_method'] != 'run out')
```

```python
def plot_bowler_map(bowler, date = None):
    if not date:
        a = df[(df.bowler_name == bowler)]
    else:
        a = df[(df.bowler_name == bowler) & (df.date == date)]
    if not date:
        plt.figure(figsize=(3,5))
        plt.scatter(a['landing_x'], 22-a['landing_y'], alpha=0.05, s=150)
        plt.scatter(a[(a.bowler_wicket==True)]['landing_x'], 22-a[(a.bowler_wicket==True)]['landing_y'], alpha=0.2, color='red', s=150)
        plt.xlim(xmin=-1, xmax=1)
        plt.ylim(ymin=0, ymax=23)
        plt.title(bowler)
        plt.show()

        currentAxis = plt.gca()
        currentAxis.add_patch(Rectangle((-0.15, 0), 0.3, 0.731, facecolor="#999999", edgecolor='k', alpha=0.5))
        plt.scatter(a.ended_x, a.ended_y, alpha = 0.07, s=150)
        plt.scatter(a[(a.bowler_wicket == True)].ended_x, a[(a.bowler_wicket == True)].ended_y, alpha = 0.3, s=150, color='red')
        plt.ylim(ymin=0, ymax=2.5)
        plt.xlim(xmin=-1, xmax=1)
        plt.title(bowler)
        plt.show()
    
    plt.figure(figsize=(8,5))
    plt.gca().add_patch(Rectangle((-0.115, 0), 0.05, 0.731, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.gca().add_patch(Rectangle((-0.002, 0), 0.05, 0.731, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.gca().add_patch(Rectangle((0.115, 0), 0.05, 0.731, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.gca().add_patch(Rectangle((-0.115, 0.711), 0.28, 0.1, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.scatter(a[(a.bowler_wicket == True)].ended_x, a[(a.bowler_wicket == True)].ended_y, alpha = 0.4, s=150, color='red')
    plt.title('Bowling heatmap of %s, red dot represents wicket taking delivery'%bowler)
    print a.control.mean()
    a = a[['ended_x', 'ended_y']].dropna()    
    plt.hist2d(a['ended_x'].values, a['ended_y'].values, bins=(15,5), cmap='Oranges', 
               range=np.array([(-1, 1), (0, 2.5)]), alpha=0.5)
    plt.gca().grid(False)
    plt.show()
```

```python
def get_batsman(batter, pace = True):
    if pace:
        a = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed > 70) & (df.dummy_ball == 1)], values='control', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        b = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed > 70) & (df.dummy_ball == 1)], values='dummy_ball', index='ball_height', columns='ball_ended_line', aggfunc=np.sum)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        c = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed > 70) & (df.dummy_ball == 1)], values='runs_batter', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        d = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed > 70) & (df.dummy_ball == 1)], values='bowler_wicket', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        return a, b, c, d
    else:
        a = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed <= 70) & (df.dummy_ball == 1)], values='control', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        b = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed <= 70) & (df.dummy_ball == 1)], values='dummy_ball', index='ball_height', columns='ball_ended_line', aggfunc=np.sum)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        c = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed <= 70) & (df.dummy_ball == 1)], values='runs_batter', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        d = pd.pivot_table(data = df[(df.batsman_name==batter) & (df.ball_speed <= 70) & (df.dummy_ball == 1)], values='bowler_wicket', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
        return a, b, c, d
```

```python
def get_bowler(bowler):
    a = pd.pivot_table(data = df[df.bowler_name==bowler], values='control', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
    b = pd.pivot_table(data = df[df.bowler_name==bowler], values='dummy_ball', index='ball_height', columns='ball_ended_line', aggfunc=np.sum)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
    c = pd.pivot_table(data = df[df.bowler_name==bowler], values='runs_batter', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
    d = pd.pivot_table(data = df[df.bowler_name==bowler], values='bowler_wicket', index='ball_height', columns='ball_ended_line', aggfunc=np.mean)[['way_outside_off', 'outside_off', 'just_outside_off', 'on_off_stump', 'on_leg_stump', 'outside_leg']].ix[['bouncer', 'high_height', 'medium_height', 'low_height']]
    return a, b, c, d
```
